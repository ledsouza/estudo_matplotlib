# Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Dados

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

Converter a coluna 'data' para datetime

```python
df['data'] = pd.to_datetime(df['data'])
```

# Análise exploratória

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_title('Temperatura no tempo', fontsize = 20)
eixo.set_ylabel('Temperatura', fontsize = 16)
eixo.set_xlabel('Data', fontsize = 16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize = 14)
```
